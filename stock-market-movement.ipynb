```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
import sklearn.metrics

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
# we will be analysing the stocks from Microsoft
msft = pd.read_csv("/kaggle/input/price-volume-data-for-all-us-stocks-etfs/Stocks/msft.us.txt")
msft.tail()
```

### Exploratory Data Analysis

```python
msft.describe()
```

### Feature Engineering

```python
# measuring the amplitude of the price variation
Amplitude = msft.High - msft.Low

# determining the direction of the stock price movement
conditions = [msft.Open < msft.Close, msft.Open == msft.Close, msft.Open > msft.Close]
choices = [1, 0, -1]
UpDown = pd.DataFrame(np.select(conditions, choices))

# mean stock price
MeanPrice = (msft.Open + msft.Close)/2
MeanPrice.head()
```

### Preprocessing the Data

```python
# features and target
feature_names = ["Open", "Close", "Low", "High", "Volume"]
features = msft[feature_names]
features["Amplitude"] = Amplitude
features["MeanPrice"] = MeanPrice

target = UpDown

# scaling
scaler = preprocessing.StandardScaler().fit(features)
scaled_X = pd.DataFrame(scaler.transform(features))

# split for the model
X_train, X_test, y_train, y_test = train_test_split(features, target)
```

### Building and Evaluating the Models

#### Random Forest Regressor

```python
rfr = RandomForestRegressor(random_state=17).fit(X_train, y_train)
preds = rfr.predict(X_test)  
```

```python
# def direction_score(true, predicted):
#     conditions = [(true < 0 & predicted < 0)]
#     choices = [1]
#     score = np.select(conditions, choices, default=1)
#     return score
```

```python
print("RANDOM FOREST SCORE: ", rfr.score(X_test, y_test))
print("MAE: ", sklearn.metrics.mean_absolute_error(y_test, preds))
print("MSE: ", sklearn.metrics.mean_squared_error(y_test, preds))
print("R2: ", sklearn.metrics.r2_score(y_test, preds))
# print("DIRECTION ERROR: ", direction_error(y_test, preds))
```

#### Random Forest Classifier

```python
rfc = RandomForestClassifier(random_state=17).fit(X_train, y_train)
preds = rfc.predict(X_test)
```

```python
print("ACCURACY: ", sklearn.metrics.accuracy_score(preds, y_test)*100, "%")
print("BALANCED ACCURACY: ", sklearn.metrics.balanced_accuracy_score(preds, y_test)*100, "%")
print("\nCLASSIFICATION REPORT\n", sklearn.metrics.classification_report(preds, y_test))
```
